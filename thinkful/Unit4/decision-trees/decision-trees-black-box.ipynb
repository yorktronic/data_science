```python
import graphlab
```

```python
# Import the Samsung data without the header -- SFrame will use the default headers instead
sf = graphlab.SFrame.read_csv('./db/samsungdata.csv', header=False, skiprows=1)
```

```python
sf
```

```python
# Rename the last two columns to 'subject' and 'activity'
# First, make sure they are the right columns
sf['X563']
sf['X564']
```

```python
sf.rename({'X563': 'subject', 'X564': 'activity'})
```

## Split the data in to training, test, and validation sets per the instructions ##

```python
# Train on subjects >= 27
sf_train= sf[sf['subject'] >= 27]
```

```python
sf_train['subject'].min()
```

```python
# Test on subjects <= 6
```

```python
sf_test = sf[sf['subject'] <= 6]
```

```python
print sf_test['subject'].max()
print sf_test['subject'].min()
```

```python
# Validate on subjects >= 21 and < 27. NOTE: I tried to use a logical filter and it wouldn't work -- this way ain't efficient
sf_validate = sf[sf.apply(lambda x: ((x['subject'] >= 21) and (x['subject'] < 27)))]
```

```python
print sf_validate['subject'].min()
print sf_validate['subject'].max()
```

## Fit a random forest with 50 estimators to the training set ##

```python
sub = round(0.08865248226950354609929078014184, 5)
print sub
```

```python
# Use column_subsample to subsample the columns for each iteration of tree construction
bb_model = graphlab.random_forest_classifier.create(sf_train, target='activity', column_subsample=sub, validation_set=sf_validate)
```

```python
# Test the RF model on the testing data
predictions = bb_model.classify(sf_test)
```

```python
results = bb_model.evaluate(sf_test)
```

```python
results
```

## The random forest classifier is nearly 80% accurate!!! ##
