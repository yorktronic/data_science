```python
import graphlab
# import pandas as pd
```

```python
sf = graphlab.SFrame('./db/samsungdata.csv')
```

```python
# Function to remove unwanted characters and substrings. Returns a cleaned column name.
cols_new_names = {}
new_column_names = []

def removeUnwanted(col_name):
    chars_to_remove = ['()', '-', ',']
    for ch in chars_to_remove:
        if ch in col_name:
            col_name = col_name.replace(ch, '')
    col_name = col_name.replace('BodyBody', 'Body')
    return col_name  
```

```python
# Populate dictionary
sf_cols = sf.column_names()
for col in sf_cols:
    cols_new_names[col] = removeUnwanted(col)

cols_new_names
```

```python
# Check to make sure the values in cols_new_names are all unique
import numpy as np

cols_new_names_values = list(cols_new_names.values())
print len(cols_new_names_values)
print len(np.unique(cols_new_names_values))
```
